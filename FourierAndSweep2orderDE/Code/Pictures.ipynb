```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(rc = {"figure.figsize":(16, 9)})
```

```python
f4p0f1 = np.loadtxt( "build/f4p0f1.txt")

net = np.linspace(0, 1, len(f4p0f1))
net2 = np.linspace(0, 1, 1000)
plt.subplot(211)
plt.scatter(net, f4p0f1, label="  ")
plt.plot(net2, - net2 * (net2 - 1) / 2, label=r" : $y = \frac{x(x-1)}{2}$")
plt.title(r" , 4 ,  $y'' = 1, \quad y(0) = y(1) = 0$")
plt.legend()
plt.subplot(212)
plt.scatter(net, f4p0f1 + net * (net - 1) / 2, label="    ", color="r")
plt.title(r"  ")
plt.legend()
# plt.plot(net2, - net2 * (net2 - 1) / 2, label=r" : $y = \frac{x(x-1)}{2}$")
plt.savefig("f4p0f1.pdf")
np.max(np.fabs(- net * (net - 1) / 2 - f4p0f1))
```

```python
s4p0f1 = np.loadtxt( "build/s4p0f1.txt")

net = np.linspace(0, 1, len(s4p0f1))
net2 = np.linspace(0, 1, 1000)
plt.subplot(211)
plt.scatter(net, s4p0f1, label="  ")
plt.plot(net2, - net2 * (net2 - 1) / 2, label=r" : $y = \frac{x(x-1)}{2}$")
plt.title(r" , 4 ,  $y'' = 1, \quad y(0) = y(1) = 0$")
plt.legend()
plt.subplot(212)
plt.scatter(net, s4p0f1 + net * (net - 1) / 2, label="    ", color="r")
plt.title(r"  ")
plt.legend()
plt.savefig("s4p0f1.pdf")
np.max(np.fabs(- net * (net - 1) / 2 - s4p0f1))
```

```python
f50p1f1 = np.loadtxt( "build/f50p1f1.txt")

def y(x):
    C = - 1 / (1 + 2.718281828459045)
    return C * np.exp(x) - (1 + C) * np.exp(- x) + 1

net = np.linspace(0, 1, len(f50p1f1))
plt.subplot(211)
plt.scatter(net, f50p1f1, label="  ")
plt.plot(net, y(net), label=r" :  $y = C e^x - (1 + C) e^{-t} + 1$,  $C = -\frac{1}{e + 1}$")
plt.title(r" , 4 ,  $y'' + y = 1, \quad y(0) = y(1) = 0$")
plt.legend()
plt.subplot(212)
plt.scatter(net, f50p1f1 - y(net), label="    ", color="r")
plt.title(r"  ")
plt.legend()
plt.savefig("f50p1f1.pdf")
np.max(np.fabs(y(net) - f50p1f1))
```
