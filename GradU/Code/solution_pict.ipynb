```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

sns.set(rc = {"figure.figsize":(16, 9)})
```

```python
solution30_900 = np.loadtxt( "build/solution.txt")

t        = solution30_900[:, 0]
x        = solution30_900[:, 1]
y        = solution30_900[:, 2]
num_ans  = solution30_900[:, 3]
anal_ans = solution30_900[:, 4]

plt.subplot(211)
plt.plot(num_ans, label=" ")
plt.plot(anal_ans,label=" ")
plt.title(r" $k=\text{const}=1$. $30$    $x$  $y$, $900$   $t$")
plt.legend()

plt.subplot(212)
plt.plot(num_ans-anal_ans, label=" (     )")
# plt.title(r" $k=\text{const}=1$")
plt.legend()
plt.savefig("constantksol.pdf")
```
