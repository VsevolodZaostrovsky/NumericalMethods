```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

sns.set(rc = {"figure.figsize":(16, 9)})
```

```python
solution30_900 = np.loadtxt( "build/solution.txt")

t        = solution30_900[:, 0]
x        = solution30_900[:, 1]
y        = solution30_900[:, 2]
num_ans  = solution30_900[:, 3]
anal_ans = solution30_900[:, 4]

plt.subplot(211)
plt.plot(num_ans, label="Численное решение")
plt.plot(anal_ans,label="Аналитическое решение")
plt.title(r"Случай $k=\text{const}=1$. $30$ точек по осям $x$ и $y$, $900$ точек по $t$")
plt.legend()

plt.subplot(212)
plt.plot(num_ans-anal_ans, label="Ошибка (разница между аналитическим и численным решениями)")
# plt.title(r"Случай $k=\text{const}=1$")
plt.legend()
plt.savefig("constantksol.pdf")
```
