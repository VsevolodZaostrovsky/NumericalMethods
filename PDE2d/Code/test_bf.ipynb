```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, pi, exp
import seaborn as sns

sns.set(rc = {"figure.figsize":(16, 9)})
```

```python
bf5_5    = np.loadtxt("build/bf5_5.txt")
bf25_5   = np.loadtxt("build/bf25_5.txt")
bf5_25   = np.loadtxt("build/bf5_25.txt")
bf10_100 = np.loadtxt("build/bf10_100.txt")
bf20_400 = np.loadtxt("build/bf20_400.txt")
```

```python
def fourier_coeff(t, n, m):
    return np.exp(- np.pi**2 *(n**2 + m**2) * t)
```

```python
net5    = np.linspace(0, 1, 5)
net25   = np.linspace(0, 1, 25)
net100   = np.linspace(0, 1, 100)
net400   = np.linspace(0, 1, 400)
net1000 = np.linspace(0, 1, 1000)

plt.plot(net5, bf5_5[0],      label=r"$N_x = 5$, $N = 5$")
plt.plot(net5, bf25_5[0],    label=r"$N_x = 25$, $N = 5$")
plt.plot(net25, bf5_25[0],    label=r"$N_x = 5$, $N = 25$")
plt.plot(net100, bf10_100[0], label=r"$N_x = 10$, $N = 100$")
plt.plot(net400, bf20_400[0], label=r"$N_x = 20$, $N = 400$")
plt.plot(net1000, fourier_coeff(net1000, 1, 1), label=r"$e^{-2 \pi ^2 t}$")
plt.title(r"     $\sin(\pi x) \sin(\pi y)$")
plt.legend()
plt.savefig("11bf_test.pdf")
```

```python
plt.plot(net5, bf5_5[1],      label=r"$N_x = 5$, $N = 5$")
plt.plot(net5, bf25_5[1],    label=r"$N_x = 25$, $N = 5$")
plt.plot(net25, bf5_25[1],    label=r"$N_x = 5$, $N = 25$")
plt.plot(net100, bf10_100[1], label=r"$N_x = 10$, $N = 100$")
plt.plot(net400, bf20_400[1], label=r"$N_x = 20$, $N = 400$")
plt.plot(net1000, fourier_coeff(net1000, 1, 1), label=r"$e^{-2 \pi ^2 t}$")
plt.title(r"     $\sin(\pi x) \sin(2\pi y)$")
plt.legend()
plt.savefig("12bf_test.pdf")
```
